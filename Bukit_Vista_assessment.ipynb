```python
import pandas as pd

book = pd.read_csv('/content/drive/My Drive/datasets/booking.csv')
book_stat = pd.read_csv('/content/drive/My Drive/datasets/booking_status.csv')
curr = pd.read_csv('/content/drive/My Drive/datasets/currency.csv')
exch = pd.read_csv('/content/drive/My Drive/datasets/exchange.csv')
```

```python
for x in book.index :
  if book.loc[x, 'booking_check_out'] < '2019-11-01':
    book.drop(x, inplace=True)

book.describe()
```

```python
for x in book.index :
  if book.loc[x,'booking_earned'] == 0:
    book.drop(x, inplace=True)

book.describe()
```

```python
for x in book.index :
  if book.loc[x, 'booking_status'] == 2 :
    book.drop(x, inplace=True)

book
```

```python
curr
```

```python
book = pd.merge(book,curr,left_on='booking_currency',right_on='currency_code')
book.head()
```

```python
book = book.drop(columns=['booking_status','booking_currency','currency_code'])
book.head()
```

```python
book.nunique()
```

```python
book_exch = pd.merge(book, exch, left_on='booking_check_in', right_on='id')
book_exch.head()
```

```python
book_exch.info()
```

```python
book_exch = book_exch[(book_exch['currency_desc'] != 'IDR')]
book_exch
```

```python
book_exch['currency_label'].nunique()
```

```python
book_exch['booking_earned'] = book_exch['booking_earned'] * book_exch['idr']
book_exch['currency_desc'], book_exch['currency_label'] = 'IDR' , 'Rp'

book_exch
```

```python
book_exch = book_exch.loc[: , ['booking_received_timestamp','booking_check_in', 'booking_check_out', 'booking_earned', 'currency_label', 'currency_desc']]
book_exch
```

```python
for x in book.index :
  if book.loc[x, 'currency_desc'] == 'USD' :
    book.drop(x, inplace=True)

book
```

```python
book = pd.concat([book, book_exch], ignore_index=True)

book
```

```python
book.reset_index()
book
```

```python
book[['booking_check_in','booking_check_out']] = book[['booking_check_in','booking_check_out']].apply(pd.to_datetime)
a = book['booking_check_out'].sub(book['booking_check_in']).dt.days
a1 = a.map(range)
b = book['booking_check_in'].add(pd.to_timedelta(a1.explode(),unit='days'))

revenue_per_day = b.to_frame('date').assign(booking_earned=book['booking_earned'].div(a)).groupby("date")['booking_earned'].sum().reset_index()

revenue_per_day
```

```python
pd.set_option('display.float_format', lambda x: '%.3f' % x)

revenue_per_day
```

```python
revenue_per_day['revenue'] = revenue_per_day['booking_earned']
revenue_per_day = revenue_per_day.drop(columns='booking_earned')
```
